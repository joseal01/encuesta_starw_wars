```python
import pandas as pd

star_wars = pd.read_csv("star_wars.csv", encoding= "ISO-8859-1")
```

 # 2: Introduction


```python
print(star_wars.columns)
star_wars.head(10)
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars
```

# 3: Cleaning Up Yes/No Columns

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts()
```

```python
yes_no = {"Yes": True,"No" : False}
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts()
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts()
```

# 4: Cleaning Up Checkbox Columns

```python
from numpy import nan

film_names = star_wars.iloc[4][3:9].tolist()
column_names = star_wars.columns[3:9]

film_names_dict =dict()

for i in range(6):
    film = film_names[i]
    dic = {film : True, nan : False}
    star_wars[column_names[i]] = star_wars[column_names[i]].map(dic)
    film_names_dict[column_names[i]]= "seen_" + str(i+1)
    
star_wars.rename(columns = film_names_dict,inplace = True)

star_wars.head()
```

# 5: Cleaning Up Ranking Columns

```python
star_wars[star_wars.columns[9:15]]= star_wars[star_wars.columns[9:15]].astype(float)

column_names = star_wars.columns[9:15]
new_names = ["ranking_"+str(i) for i in range(1,7)]

ranking_dict = dict(zip(column_names,new_names))

star_wars.rename(columns=ranking_dict,inplace = True)

star_wars.columns[9:15]



```

# 6: Finding The Highest Ranked Movie

```python
%matplotlib inline
import numpy as np

star_wars[star_wars.columns[9:15]].apply(np.mean).plot.bar()
```

- Primero he convertido los si y no de dos columnas en booleanos para que se puedan tratar mejor.
- Despues he hecho lo mismo con las columnas en las que se preuguntaba si habian visto alguna de las peliculas, y ademas he renombrado esas columnas para hacer mas facil la lectura de los datos.
- Por ultimo he convertido los datos del ranking a float y renombrado las columnas de los rankings
- Al final el orden quedaria(de mejor a peor) : 5>6>4>1>2>3. Es evidente que las tres primeras peliculas(en tiempo) son las que mejor acogida tienen, mientras que las mas recientes estan peor valoradas. 


```python
star_wars[star_wars.columns[3:9]].apply(np.sum).plot.bar()
```

- Existe una correlacion negativa entre las dos. La unica que "emperoa" la correlacion es la primera, que a pesar de estar peor valorada que la 4, fue mas vista. Esto seguramente sea debido a que la 4 fue la primera pelicula en salir, unido al hecho que el episodio 1 genero mucha expectacion al ser la primera de star wars en mas de 20 años. 

```python
import matplotlib.pyplot as plt
true_fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]==True]
not_true_fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] ==False]

fig , ax = plt.subplots(2,2, figsize = (20,20))

true_fan[true_fan.columns[9:15]].apply(np.mean).plot.bar(ax=ax[0][0])
not_true_fan[true_fan.columns[9:15]].apply(np.mean).plot.bar(ax=ax[0][1])
true_fan[true_fan.columns[3:9]].apply(np.sum).plot.bar(ax=ax[1][0])
not_true_fan[true_fan.columns[3:9]].apply(np.sum).plot.bar(ax=ax[1][1])

ax[0][0].set_title("True Fan Rate-Mean")
ax[0][1].set_title("Not True Fan Rate-Mean")
ax[1][0].set_title("True Fan Seen")
ax[1][1].set_title("Not True Fan Seen")
```

- A la hora de evaluar hay una clara diferencia entre los fans y no fans. Sin embargo, siguen tendencias parecidas, en cuanto creciemineto y decrecimiento en el del numero de peliculas vistas.
